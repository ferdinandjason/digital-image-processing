# Segmentation based on Histogram

```python
import numpy as np
import matplotlib.pyplot as plt

from skimage import io, color, draw, filters, segmentation, util
```

```python
image_text = color.rgb2gray(io.imread('./segmentasi.png'))
plt.figure(figsize=(12,6))
plt.imshow(image_text, cmap='gray')
plt.axis('off')
```

```python
plt.hist(image_text.flatten() * 256, bins=32, range=(0,256))
plt.xlim(0,256)
```

### Global Thresholding

```python
threshold = 120
plt.figure(figsize=(12,6))
plt.imshow(image_text > threshold/255, cmap='gray')
plt.axis('off')
```

### Threshold Otsu

```python
image_otsu = filters.threshold_otsu(image_text)
plt.figure(figsize=(12,6))
plt.imshow(image_text > image_otsu, cmap='gray')
plt.axis('off')
```

### Threshold Local

```python
image_local = filters.threshold_local(image_text, 19, offset=0.04)
plt.figure(figsize=(12,6))
plt.imshow(image_text > image_local, cmap='gray')
plt.axis('off')
```

### Filter Sobel

```python
image_sobel = filters.sobel(image_text)
plt.figure(figsize=(12,6))
plt.imshow(image_sobel, cmap='gray')
plt.axis('off')
```

### Filter Laplace

```python
image_laplace = filters.laplace(image_text)
plt.figure(figsize=(12,6))
plt.imshow(image_laplace, cmap='gray')
plt.axis('off')
```

### Sobel + Threshold Local

```python
image_sobel2 = filters.sobel(image_text)
image_sobel2 = util.invert(image_sobel2)
image_local_sobel2 = filters.threshold_local(image_sobel2, 21, offset=0.03)
plt.figure(figsize=(12,6))
plt.imshow(image_sobel2 > image_local_sobel2, cmap='gray')
plt.axis('off')
```
