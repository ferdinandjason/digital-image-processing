# Image Denoising by FFT

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
image = mpimg.imread('./moonlanding.png').astype(float)

plt.figure(figsize=(10,10))
plt.imshow(image)
plt.title('Original Image')
plt.set_cmap('gray')
```

Compute the 2D FFT of the input image and plot the result

```python
image_fft = np.fft.fft2(image)

plt.figure(figsize=(10,10))
plt.imshow(np.abs(np.log10(image_fft)))
plt.title('FFT Spectrum')
plt.set_cmap('gray')
```

Filtering spektrum fft dengan men-truncate coeffisien

```python
keep_fraction = 0.08

image_fft2 = image_fft.copy()
r,c = image_fft2.shape
image_fft2[int(r*keep_fraction):int(r*(1-keep_fraction))] = 0
image_fft2[:, int(c*keep_fraction):int(c*(1-keep_fraction))] = 0

plt.figure(figsize=(10,10))
plt.imshow(np.abs(np.log10(image_fft2)))
plt.title('FFT Spectrum Truncated Version')
plt.set_cmap('gray')
```

Reconstruct the final image

```python
image_reconstructed = np.fft.ifft2(image_fft2).real

plt.figure(figsize=(10,10))
plt.imshow(image_reconstructed)
plt.title('Reconstructed Image')
plt.set_cmap('gray')
```
