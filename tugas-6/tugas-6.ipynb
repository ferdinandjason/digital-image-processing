```python
import cv2
import skimage
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('./daun.jpg',cv2.IMREAD_GRAYSCALE)
```

```python
ret,threshold_image = cv2.threshold(image,200,255,cv2.THRESH_BINARY_INV)
plt.imshow(threshold_image, cmap='gray')
```

```python
from skimage.morphology import thin

thin_daun = thin(threshold_image)
plt.imshow(thin_daun, cmap='gray')
```
